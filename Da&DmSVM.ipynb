```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer
import matplotlib.pyplot as plt
import seaborn as sns

import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.stats.stattools import durbin_watson
```

```python
df = pd.read_csv("/content/water-quality-FINAL-cleaned.csv")
```

```python
#MLR analysis
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.stats.stattools import durbin_watson
```

```python
target = 'Temperature'
predictors = [
    'pH', 'DO', 'Conductivity',
    'Total_N', 'Orthophosphate']
```

```python
df = df.dropna()
df["WQI"] = (
    df['pH'] * .17 +
    df['DO'] * .26 +
    df['Conductivity'] * .11 +
    df['Temperature'] * .15 +
    df['Total_N'] * .15 +
    df['Orthophosphate'] * .15
)
```

```python
from sklearn import svm
X = df[['pH', 'DO', 'Conductivity', 'Temperature', 'Total_N', 'Orthophosphate']]
df['WQI_bins'] = pd.cut(df['WQI'], bins = [0, .50, .70, 1.00], labels = ['Poor', 'Fair', 'Good'])
y = df['WQI_bins']
clf = svm.SVC()
clf.fit(X, y)
```

```python
clf.predict([[7.5, 8.0, 200, 15, 1.2, 0.1]])
```

```python
# Help used: https://scikit-learn.org/stable/modules/svm.html

from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn import svm
X_train , X_test , y_train, y_test = train_test_split(X, y, random_state=0)

clf = svm.SVC()
clf.fit(X_train, y_train)

clf.score(X_test, y_test)
```

```python
y.value_counts()
```

```python
# Help used: https://scikit-learn.org/stable/modules/generated/sklearn.metrics.confusion_matrix.html
# Help used: https://scikit-learn.org/stable/modules/generated/sklearn.metrics.classification_report.html
# Help used: https://scikit-learn.org/stable/modules/generated/sklearn.metrics.accuracy_score.html

from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

y_pred = clf.predict(X_test)

print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# Help used: https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVR.html

from sklearn.svm import SVR
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error

y2 = df['WQI']

X_train , X_test , y_train, y_test = train_test_split(X, y2, random_state=0)

regr = make_pipeline(StandardScaler(), SVR(C=1.0, epsilon=0.2))
regr.fit(X_train, y_train)

y_pred = regr.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print("MAE: ", mae)
print("MSE: ", mse)
```
